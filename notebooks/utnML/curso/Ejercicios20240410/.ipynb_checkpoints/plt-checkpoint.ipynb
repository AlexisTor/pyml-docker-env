```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1024
np.random.seed(1234)
x = np.random.normal(0,1,n)
y = np.random.normal(0,1,n)
```

```python
plt.scatter(x,y, alpha=0.3, edgecolors='none')
plt.axis('off')
```

```python
color = -(x**2 + y**2)
```

```python
plt.scatter(x,y, c=color, cmap='jet', alpha=0.3, edgecolors='none')
plt.axis('off')
```

```python
scale = 100 * np.random.rand(n)
```

```python
plt.scatter(x,y, c=color, cmap='jet', s=scale, alpha=0.3, edgecolors='none')
plt.axis('off')
```
